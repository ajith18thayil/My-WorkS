```python
X = dataset.iloc[:,2:32].values
y = dataset.iloc[:,1].values
```

```python
y.shape
```

```python
from sklearn.preprocessing import LabelEncoder

labelEncoder=LabelEncoder()
```

```python
y =labelEncoder.fit_transform(y)
```

```python
y
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.1,random_state = 42)
```

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()

X_train = sc.fit_transform(X_train)
X_test = sc.fit_transform(X_test)
```

```python
##ANN
```

```python
import keras
from keras.models import Sequential
from keras.layers import Dense
```

```python
#initialising ANN
classifier = Sequential()
```

```python

# units = (x+1)/2
#       = (30+1)/2
#       = 16
```

```python
#creating input layer and first hidden layer
classifier.add(Dense(units = 16,kernel_initializer='uniform',activation = 'sigmoid',input_dim = 30))

#second hidden layer
classifier.add(Dense(units = 16,kernel_initializer='uniform',activation = 'sigmoid'))


#output layer
classifier.add(Dense(units = 1,kernel_initializer='uniform',activation = 'relu'))
```

```python
classifier.compile(optimizer = 'adam',loss = 'binary_crossentropy',metrics =['accuracy'])
```

```python
classifier.fit(X_train,y_train,batch_size = 10,epochs = 100)
```

```python
y_pred = classifier.predict(X_test)
y_pred = (y_pred>0.5)
```

```python
y_pred
```

```python
from sklearn.metrics import confusion_matrix

cm= confusion_matrix(y_test,y_pred)
```
