```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img=cv2.imread(r'C:\Users\HP\OneDrive\Desktop\Ajith\ajith new pic.jpg')
```

```python
img.shape
```

```python
cv2.imshow('group',img)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
plt.imshow(img);
```

```python
#BGR to RGB
con_img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(con_img);
```

```python
# Gray -Scale 

grey=cv2.cvtColor(con_img,cv2.COLOR_RGB2GRAY)
plt.imshow(grey);
```

```python
cv2.imshow('grey',grey)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
#Binary

s,bin_img=cv2.threshold(grey,128,255,cv2.THRESH_BINARY|cv2.THRESH_OTSU)

cv2.imshow('binary_img',bin_img)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
plt.imshow(bin_img)
```

```python
# edge detection

ed_img=cv2.Canny(grey,150,200)
plt.imshow(ed_img)
```

```python
#Drawing shapes and text mentioning

con_img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)

s_img=cv2.rectangle(con_img,(110,80),(200,190),(0,255,0),2)
s_img=cv2.rectangle(con_img,(300,70),(400,180),(0,255,0),2)
s_img=cv2.rectangle(con_img,(530,130),(640,250),(0,255,0),2)
s_img=cv2.putText(con_img,'Person-1',(65,60),cv2.FONT_HERSHEY_COMPLEX,1,(255, 255, 255),thickness=4)
s_img=cv2.putText(con_img,'Person-2',(280,30),cv2.FONT_HERSHEY_COMPLEX,1,(255, 255, 255),thickness=4)
s_img=cv2.putText(con_img,'Person-3',(510,120),cv2.FONT_HERSHEY_COMPLEX,1,(255, 255, 255),thickness=4)

plt.imshow(s_img)
```
