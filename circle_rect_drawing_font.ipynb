```python
img_grey=cv2.imread(r'ajith new pic.jpg',cv2.IMREAD_GRAYSCALE)
```

```python
img_grey.shape
```

```python
cv2.imshow('grayscale image',img_grey)
cv2.imshow('original image',img)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
(thresh ,binary_img)=cv2.threshold(img_grey,128,255,cv2.THRESH_BINARY | cv2.THRESH_OTSU)
```

```python
cv2.imshow('binary image',binary_img)
cv2.imshow('original image', img)
cv2.imshow('greyscale image',img_grey)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
edge_img=cv2.Canny(img,255,0)
```

```python
cv2.imshow('orginal image', img)
cv2.imshow('edge image',edge_img)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
import matplotlib.pyplot as plt
```

```python
plt.imshow(img)
```

```python
converted_img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(converted_img)
```

```python
converted_img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
rect=cv2.rectangle(converted_img,pt1=(1000,50),pt2=(850,300),color=(0,255,255),thickness=10)
plt.imshow(rect)
```

```python
converted_img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
circle_img=cv2.circle(converted_img,center=(300,300),radius=100,color=(255,255,0),thickness=-3)

plt.imshow(circle_img)
```

```python
converted_img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)

text_img=cv2.putText(converted_img,text='luminar',org=(100,300),fontFace=cv2.FONT_HERSHEY_PLAIN,fontScale=5,color=(255,0,0),thickness=4)

plt.imshow(text_img)
```
