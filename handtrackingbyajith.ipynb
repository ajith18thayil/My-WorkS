```python
import cv2
import mediapipe as mp
import time
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread(r'C:\Users\AJITH\Videos\computer vision\open cv\hand type fingure.jpg')
```

```python
plt.imshow(img)
```

```python
converted = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
cv2.imshow('hand',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
mp_hands = mp.solutions.hands
```

```python
hands = mp_hands.Hands(static_image_mode=True,max_num_hands=2,min_detection_confidence=0.3)
```

```python
mp_drawing = mp.solutions.drawing_utils
```

```python
results=hands.process(converted)
```

```python
results
```

```python
if results.multi_hand_landmarks:
    for hand_no,hand_landmarks in enumerate(results.multi_hand_landmarks):

        mp_drawing.draw_landmarks(image= img,landmark_list=hand_landmarks,connections= mp_hands.HAND_CONNECTIONS)
    cv2.imshow('drawing image',img)
    cv2.waitKey(0)
    cv2.destroyAllWindows()

```

```python
cv2.imshow('drawing image',img)
cv2.waitKey(5)
cv2.destroyAllWindows()
```
