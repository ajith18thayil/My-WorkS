```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('/content/CC GENERAL.csv',nrows=1000)
df
```

```python
df.shape
```

```python
df.isna().sum()
```

```python
df.info()
```

```python
df.fillna(df['MINIMUM_PAYMENTS'].mean(),inplace=True)
```

```python
df.isna().sum()
```

```python
df.drop(['CUST_ID'],axis=1,inplace=True)
df
```

```python
from sklearn.preprocessing import normalize
data_scaled=normalize(df)
data_scaled
```

```python
df_new=pd.DataFrame(data_scaled,columns=df.columns)
df_new
```

```python
import scipy.cluster.hierarchy as shc
import matplotlib.pyplot as plt
plt.figure(figsize=(10,10))
dend=shc.dendrogram(shc.linkage(df_new,method='complete'))
plt.title('Dendrogram')
```

```python
from sklearn.cluster import AgglomerativeClustering
cluster=AgglomerativeClustering(n_clusters=3,affinity='euclidean',linkage='complete')
y_pred=cluster.fit_predict(df_new)
```

```python
# df_new['cluster']=y_pred
# df_new
```

```python
plt.figure(figsize=(10,10))
plt.scatter(df_new['PAYMENTS'],df_new['PURCHASES'],c=cluster.labels_)
```

```python
from sklearn.cluster import KMeans
wcss=[]
for i in range(1,11):
  km=KMeans(n_clusters=i,init='k-means++',random_state=42)
  km.fit(df_new)
  wcss.append(km.inertia_)
```

```python
import matplotlib.pyplot as plt
plt.plot(range(1,11),wcss)
plt.xlabel('Number of clusters')
plt.ylabel=('WCSS')
plt.show()
```

```python
km=KMeans(n_clusters=5,init='k-means++',random_state=42)
km.fit(df_new)
y_pred_km=km.predict(df_new)
y_pred_km
```

```python
df_new['cluster']=y_pred_km
df_new
```

```python
import seaborn as sns
sns.pairplot(df_new,hue='cluster')
```
