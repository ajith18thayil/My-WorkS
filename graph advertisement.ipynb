```python

import matplotlib.pyplot as plt
from matplotlib.patches import FancyBboxPatch

# Set up the plot
fig, ax = plt.subplots(figsize=(10, 8))
ax.axis('off')

# Define box properties
box_props = dict(boxstyle="round,pad=0.3", edgecolor="black", facecolor="#add8e6")

# Define text properties
text_props = dict(ha='center', va='center', fontsize=12)

# Define flowchart boxes and connections
elements = [
    ("Video Feed & User Interaction Data", 0.5, 0.9),
    ("Preprocessing", 0.5, 0.8),
    ("Face Detection & Recognition\nDeepFace, FaceNet", 0.3, 0.7),
    ("Demographic Analysis\nViT, Swin Transformer", 0.7, 0.7),
    ("NLP - Text Analysis\nBERT, GPT-4", 0.5, 0.6),
    ("Multimodal Analysis\nCLIP", 0.5, 0.5),
    ("Ad Content Selection", 0.5, 0.4),
    ("Custom Content Creation\nStyleGAN3", 0.3, 0.3),
    ("Real-Time Data Processing\nYOLOv5", 0.7, 0.3),
    ("Performance Monitoring", 0.5, 0.2),
    ("CCTV Monitoring", 0.3, 0.1),
    ("Real-Time Analytics Dashboard", 0.7, 0.1)
]

# Draw the flowchart elements
for text, x, y in elements:
    ax.text(x, y, text, bbox=box_props, **text_props)

# Draw arrows connecting the boxes
arrow_props = dict(facecolor='black', arrowstyle='->', connectionstyle="arc3,rad=0.3")

connections = [
    (0.5, 0.87, 0.5, 0.82),  # Input to Preprocessing
    (0.5, 0.77, 0.3, 0.72),  # Preprocessing to Face Detection
    (0.5, 0.77, 0.7, 0.72),  # Preprocessing to Demographic Analysis
    (0.3, 0.67, 0.5, 0.62),  # Face Detection to NLP
    (0.7, 0.67, 0.5, 0.62),  # Demographic Analysis to NLP
    (0.5, 0.57, 0.5, 0.52),  # NLP to Multimodal Analysis
    (0.5, 0.47, 0.5, 0.42),  # Multimodal Analysis to Ad Content Selection
    (0.5, 0.37, 0.3, 0.32),  # Ad Content Selection to Custom Content Creation
    (0.5, 0.37, 0.7, 0.32),  # Ad Content Selection to Real-Time Processing
    (0.5, 0.27, 0.5, 0.22),  # Ad Content Selection to Performance Monitoring
    (0.5, 0.17, 0.3, 0.12),  # Performance Monitoring to CCTV Monitoring
    (0.5, 0.17, 0.7, 0.12)   # Performance Monitoring to Analytics Dashboard
]

for (x1, y1, x2, y2) in connections:
    ax.annotate("", xy=(x2, y2), xytext=(x1, y1), arrowprops=arrow_props)

# Display the flowchart
plt.show()
```
